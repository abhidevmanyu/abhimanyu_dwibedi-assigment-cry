1) Implement Affine Cipher.

```python
def gcd(a, b):
    while b != 0:
        a, b = b, a % b
    return a

def mod_inverse(a, m):
    if gcd(a, m) != 1:
        return None
    x, y, u, v = 0, 1, 1, 0
    while a != 0:
        q, r = m // a, m % a
        m, a = a, r
        x, y, u, v = u, v, x - q * u, y - q * v
    return x % m

def affine_encrypt(plain_text, a, b):
    encrypted_text = ""
    for char in plain_text:
        if char.isalpha():
            # Apply the affine transformation
            encrypted_char = chr((a * (ord(char) - ord('A')) + b) % 26 + ord('A'))
            encrypted_text += encrypted_char
        else:
            encrypted_text += char
    return encrypted_text

def affine_decrypt(cipher_text, a, b):
    decrypted_text = ""
    a_inverse = mod_inverse(a, 26)
    if a_inverse is None:
        return "Error: 'a' is not coprime with 26."
    for char in cipher_text:
        if char.isalpha():
            # Apply the inverse affine transformation
            decrypted_char = chr(a_inverse * (ord(char) - ord('A') - b) % 26 + ord('A'))
            decrypted_text += decrypted_char
        else:
            decrypted_text += char
    return decrypted_text

# Example usage
plain_text = "HELLO WORLD"
a = 5
b = 8

cipher_text = affine_encrypt(plain_text, a, b)
print("Cipher Text:", cipher_text)

decrypted_text = affine_decrypt(cipher_text, a, b)
print("Decrypted Text:", decrypted_text)
```

Implement Extended Euclidean Algorithm.

```python
def extended_gcd(a, b):
    if b == 0:
        return a, 1, 0
    gcd, x1, y1 = extended_gcd(b, a % b)
    x = y1
    y = x1 - (a // b) * y1
    return gcd, x, y

# Example usage
a = 48
b = 18

gcd, x, y = extended_gcd(a, b)
print("GCD:", gcd)
print("x:", x)
print("y:", y)
```

```python

import hashlib

def compute_hash(message):
    md5_hash = hashlib.md5()
    md5_hash.update(message.encode('utf-8'))
    return md5_hash.hexdigest()

# Original Message
original_message = "Hello, World!"

# Compute hash of the original message
original_hash = compute_hash(original_message)
print("Original Message:", original_message)
print("Original Hash:", original_hash)

# Simulate the receiver computing the hash again and ensuring its integrity
received_message = original_message
received_hash = compute_hash(received_message)
print("Received Message:", received_message)
print("Received Hash:", received_hash)
print("Integrity Check:", original_hash == received_hash)

# Change the message slightly
changed_message = "Hello, World!!"
changed_hash = compute_hash(changed_message)
print("Changed Message:", changed_message)
print("Changed Hash:", changed_hash)
print("Integrity Check:", original_hash == changed_hash)
```

. a)  Create a password file of 10 passwords and use it for identification.
    b)  Modify one to store the hash values of passwords & use it.
    c)  Optional: Create a salt file; add salt to password before storing in (b). 


```python
import hashlib

# Function to compute the MD5 hash of a given string
def compute_hash(str):
    md5_hash = hashlib.md5()
    md5_hash.update(str.encode('utf-8'))
    return md5_hash.hexdigest()

# Step 1: Create a password file
passwords = ["password1", "123456", "qwerty", "letmein", "hello123",
             "abc123", "pass1234", "password123", "admin", "welcome"]

with open("passwords.txt", "w") as password_file:
    password_file.write('\n'.join(passwords))

print("Password file created successfully.")

# Step 2: Modify the file to store the hash values of passwords
with open("passwords.txt", "r") as password_file:
    with open("hashed_passwords.txt", "w") as hashed_password_file:
        salt = "somesalt"  # Optional: Salt value

        for password in password_file:
            password = password.strip()
            salted_password = salt + password
            hashed_password = compute_hash(salted_password)
            hashed_password_file.write(hashed_password + '\n')

print("Hashed password file created successfully.")

# Step 3: Optional - Create a salt file
with open("salt.txt", "w") as salt_file:
    salt_file.write(salt)

print("Salt file created successfully.")
```

Generate a symmetric key by using Diffie-Hellman method given g, p, x and y

```python
def generate_symmetric_key(g, p, x, y):
    # Calculate public keys for Party A and Party B
    public_key_a = pow(g, x, p)
    public_key_b = pow(g, y, p)

    # Calculate the shared secret key for Party A and Party B
    shared_secret_a = pow(public_key_b, x, p)
    shared_secret_b = pow(public_key_a, y, p)

    # Verify that both shared secret keys are the same
    assert shared_secret_a == shared_secret_b

    # Return the symmetric key (shared secret)
    return shared_secret_a

# Parameters
g = 2  # Generator
p = 23  # Prime number
x = 6  # Private key for Party A
y = 15  # Private key for Party B

# Generate symmetric key
symmetric_key = generate_symmetric_key(g, p, x, y)
print("Symmetric Key:", symmetric_key)
```
